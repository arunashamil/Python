```python
import pandas as pd
```

```python
excel = pd.read_excel('students_info.xlsx')
```

```python
ejudge = pd.read_html('results_ejudge.html')[0]
```

```python
excel
```

```python
ejudge
```

```python
data = pd.merge(excel, ejudge, how = 'left', right_on = 'User', left_on = 'login')
```

```python
data
```

```python
data.fillna(0)
```

```python
def hey_fac(df):
    if (df.G + df.H >= 10.0):
        return df.group_faculty
newcomers = data.fillna(0).apply(hey_fac, axis = 'columns')
newcomers
```

```python
newcomers.fillna(0)
```

```python
for x in newcomers.fillna(0):
    if x != 0:
        print(str(int(x)) + ' ')
```

```python
def hey_out(df):
    if (df.G + df.H >= 10.0):
        return df.group_out
alreadycomers = data.fillna(0).apply(hey_out, axis = 'columns')
alreadycomers
```

```python
for x in newcomers.fillna(0):
    for y in alreadycomers.fillna(0):
        if (x != 0) & (y != 0):
            print(str(int(x)) + ' -> ' + str(int(y)))
```
